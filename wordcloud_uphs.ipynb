Looping for running the same code to generate the wordcloud in all of the txt files from each UPH

```python
#Defining the names of each file 
uphs = ['uph_5','uph_6','uph_7','uph_8']
for uph in uphs:
    text = open(uph+'.txt', mode = 'r', encoding = 'utf-8').read() #I have assembled a txt file with all the contributions 
    #here setting the parameters for wordcloud                     # for each UPH (from spreadsheet) named after them
    wc = WordCloud(                                               
    background_color = 'white',
    max_font_size=40,   
    stopwords = stopwords)
    wc.generate(text) #generating wordcloud according to the txt file I opened (which can also be an abstract, keywords, etc)
    wc.to_file('wordcloud_'+uph+'.png') #Saving the wordclouds to my directory
#plotting the last one, outside of loop 
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
```
